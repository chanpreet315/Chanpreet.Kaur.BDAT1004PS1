Which of the following words is the longest:
Supercalifragilisticexpialidocious, Honorificabilitudinitatibus, or
Bababadalgharaghtakamminarronnkonn? 

```
string[] words = { "Supercalifragilisticexpialidocious", "Honorificabilitudinitatibus", "Bababadalgharaghtakamminarronnkonn" };

int maxLength = 0;
List<string> longestWords = new List<string>();

foreach (string word in words)
{
    int length = word.Length;
    if (length > maxLength)
    {
        longestWords.Clear();
        longestWords.Add(word);
        maxLength = length;
    }
    else if (length == maxLength)
    {
        longestWords.Add(word);
    }
}

Console.WriteLine("Longest word(s) with length " + maxLength + ":");
foreach (string longestWord in longestWords)
{
    Console.WriteLine(longestWord);
}
```

# Question 3 C#


Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c)
, where
s = (a+b+c)/2.
>>> triangleArea(2,2,2)
1.7320508075688772


```
public void triangleArea(int a, int b, int c)
{
    int s = (a + b + c) / 2;
    double calculation = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
    Console.WriteLine($"Area of triangle ({a},{b},{c}) is: {calculation}");
}
triangleArea(2, 2, 2);
```

# Question 4 C#

Write a program in C# Sharp to separate odd and even integers in separate arrays.
Go to the editor
Test Data :
Input the number of elements to be stored in the array :5
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32
Expected Output:
The Even elements are:
42 56 32
The Odd elements are :
25 47

using System;

namespace SeparateOddEven
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.Write("Input the number of elements to be stored in the array: ");
            int numOfElements = Convert.ToInt32(Console.ReadLine());
            int[] elements = new int[numOfElements];
            Console.WriteLine("Input " + numOfElements + " elements in the array: ");

            for (int i = 0; i < numOfElements; i++)
            {
                Console.Write("element - " + i + " : ");
                elements[i] = Convert.ToInt32(Console.ReadLine());
            }

            int[] evenElements = Array.FindAll(elements, x => x % 2 == 0);
            int[] oddElements = Array.FindAll(elements, x => x % 2 != 0);

            Console.WriteLine("The Even elements are: " + string.Join(", ", evenElements));
            Console.WriteLine("The Odd elements are: " + string.Join(", ", oddElements));

            Console.ReadLine();
        }
    }
}
